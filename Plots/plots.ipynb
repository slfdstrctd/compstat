```r
library(ggplot2)
library(dplyr)

data(mtcars, package="datasets")
```

## 1. Продемонстрируйте зависимость потребления топлива от прочих характеристик

```r
# mtcars %>% 
#   select(mpg, cyl, disp, hp, drat, wt, qsec, vs, am, gear, carb) %>% 
#   explore_all(target = mpg)

# cyl
ggplot(mtcars, aes(x=mpg, y=as.factor(cyl))) + 
    geom_boxplot(fill="lightblue", alpha=0.5) + 
    theme_bw() + xlab("Miles/gallon") + ylab("Cylinders")

# disp
ggplot(mtcars, aes(x=mpg, y=disp)) + 
    geom_point(aes(size=2), colour = "blue", alpha = 0.5) +
    xlab("Miles/gallon") + ylab("Displacement (cu.in.)")

# drat
ggplot(mtcars, aes(x=mpg, y=drat)) + 
    geom_point(aes(size=1.5), colour = "red", alpha = 0.5) +
    geom_smooth(formula = "y~x", method = "lm", se = FALSE, color = "black") +
    xlab("Miles/gallon") + ylab("Rear axle ratio")

# hp
ggplot(mtcars, aes(x=mpg, y=hp)) + geom_point(alpha = 0.5) + 
    xlab("Miles/gallon") + ylab("Horsepower")

# wt
ggplot(mtcars, aes(x=mpg, y=wt)) + 
   geom_point(colour = "brown", alpha = 0.5) +
   geom_smooth(formula = "y~x", method = "lm", se = FALSE, color = "black")+
   xlab("Miles/gallon") + ylab("Weight (lb/1000)")

# qsec
ggplot(mtcars, aes(x=mpg, y=qsec)) + geom_point(colour = "orange", alpha = 0.5) +
    xlab("Miles/gallon") + ylab("1/4 mile time")

# vs
ggplot(data = mtcars, aes(x = mpg, color = as.factor(vs))) +
  geom_density() + xlab("Miles/gallon") + ylab("V/S") +
  scale_color_discrete("V/S", labels = c("V-shape", "Straight"))

# am 
ggplot(data = mtcars, aes(x = mpg, fill = as.factor(am))) +
  geom_density() + xlab("Miles/gallon") + ylab("Transmission") + 
  scale_fill_discrete("Transmission", labels = c("auto", "manual"))

# gear
ggplot(mtcars, aes(x=mpg, y=as.factor(gear))) + 
    geom_boxplot(fill="lightgreen", alpha=0.5) + 
    theme_bw() + xlab("Miles/gallon") + ylab("Forward gears")
    
# carb
ggplot(mtcars, aes(x=mpg, y=as.factor(carb))) + 
    geom_boxplot(fill="orange", alpha=0.5) + 
    theme_bw() + xlab("Miles/gallon") + ylab("Carburetors")
```

## 2. Покажите зависимость мощности двигателя от его объема, типа коробки передач и т.п.

```r
# mtcars %>% 
#   select(hp, disp, am, gear, carb) %>% 
#   explore_all(target = hp)

# disp
ggplot(mtcars, aes(x=hp, y=disp)) + 
    geom_point(colour = "blue", alpha = 0.5) +
    geom_smooth(formula = "y~x", method = "loess", color = "black") +
    xlab("Horsepower") + ylab("Displacement (cu.in.)")

# am
ggplot(mtcars, aes(x=hp, y=as.factor(am))) +
  geom_bar(stat="identity", width=0.5, fill="red", alpha=0.5) + 
  theme_bw() + xlab("Horsepower") + ylab("Transmission")

# gear
ggplot(mtcars, aes(x=hp, y=as.factor(gear))) + 
    geom_boxplot(fill="lightblue", alpha=0.5) + 
    theme_bw() + xlab("Horsepower") + ylab("Forward gears")

# carb
ggplot(mtcars, aes(x=hp, y = as.factor(carb))) + 
    geom_boxplot(fill="lightgreen", alpha=0.5) + 
    theme_bw() + xlab("Horsepower") + ylab("Carburetors")
```
